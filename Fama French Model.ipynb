```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
fff = pd.read_excel('Desktop/IC/Term 1/IPM/data.xls', index_col=0)
fff.head()
```

```python
# excess returns
index = 0
for index in range(4, 14):
    fff[f'Excess_{index-3}'] = fff.iloc[:, index]-fff['Risk-Free']
fff.head()
```

```python
X = sm.add_constant(fff.loc["19900131":"20021231", ['Mkt (vwretd)', 'SMB', 'HML']])
model = sm.OLS(fff.loc["19900131":"20021231", ['Excess_1']], X)
results = model.fit()
print(results.summary())
```

```python
beta_mkt = pd.Series(name='beta_mkt')
beta_smb = pd.Series(name='beta_smb')
beta_hml = pd.Series(name='beta_hml')

for index in range(1, 11):
    # Perform the OLS regression
    model = sm.OLS(fff[f'Excess_{index}'], X)
    results = model.fit()
    
    # Collect the estimated coefficients
    beta_mkt.loc[f'CAP{index}RET'] = results.params[1]
    beta_smb.loc[f'CAP{index}RET'] = results.params[2]
    beta_hml.loc[f'CAP{index}RET'] = results.params[3]
```

```python
mpr = pd.read_excel('Desktop/IC/Term 1/IPM/mpr.xls', index_col=0)
mpr.head()
```

```python
returnex_avg = mpr.iloc[:, 0:10].mean()
returnex_avg
```

```python
X1 = pd.concat([beta_mkt, beta_smb, beta_hml], axis=1)
X1 = sm.add_constant(X1, prepend=True)
#X1 = sm.add_constant(pd.concat([beta_mkt, beta_smb, beta_hml], axis=1))
```

```python
model = sm.OLS(returnex_avg, X1)
results = model.fit()
print(results.summary())
```
